### 2. Simulation

#### Simulation de la loi ùëã~ùê∫ùëíùëú(ùúÜ) avec ùúÜ = 0.3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# theorie
lambda_x = 0.3  

# loi geometrique X
k = np.arange(1, 15)
geoX =  (1-lambda_x)**(k-1)*lambda_x     
plt.bar(k, geoX, color="red", label="Th√©orique")

# empirique
essais = 1000
expX =[]
for i in range(essais) :
    cpt = 0
    while np.random.rand() > lambda_x : 
        cpt += 1
    expX.append(cpt + 1)
    
# Affichage
plt.hist(expX, bins=np.arange(1, np.max(expX)+2) - 0.5, density = True, alpha = 0.6, label="Empirique") 
plt.legend()
plt.show()

# Conclusion
print("L'esp√©rance math theorique de X est", 1/lambda_x)
print("L'esp√©rance math empirique de X est", np.mean(expX))
print("L'ecart type thoerique de X est", np.sqrt((1-lambda_x)/lambda_x**2))
print("L'ecart type empirique de X est", np.std(expX))
```

#### Simulation de ùëå~ùê∫ùëíùëú(ùúá)

```python
# theorie
micro_y = 0.5 

# loi geometrique Y
k = np.arange(1, 10)
geoY =  (1-micro_y)**(k-1)*micro_y      
plt.bar(k, geoY, color="red", label="Th√©orique")

# experimentation
essais = 100
expY =[]
for i in range(essais) :
    cpt = 0
    while np.random.rand() > micro_y : 
        cpt += 1
    expY.append(cpt + 1)

# Affichage
plt.hist(expY, bins=np.arange(1, np.max(expY)+2) - 0.5, density = True, alpha = 0.6, label="Empirique") 
plt.legend()
plt.show()

# Conclusion
print("L'esp√©rance math theorique de X est", 1/micro_y)
print("L'esp√©rance math empirique de X est", np.mean(expY))
print("L'ecart type thoerique de X est", np.sqrt((1-micro_y)/micro_y**2))
print("L'ecart type empirique de X est", np.std(expY))
```

```python
# theorie
p_z = 1-(1-lambda_x)*(1-micro_y)     #calcul theorique manuel

# loi geometrique X
k = np.arange(1, 7)
geoZ = (1-p_z)**(k-1)*p_z      
plt.bar(k, geoZ, color="red", label="Th√©orique loi de poisson")

# experimentation
essais = 100
expZ =[]
for i in range(essais) :
    cpt = 0
    while np.random.rand() > p_z : 
        cpt += 1
    expZ.append(cpt + 1)
    
# Affichage
plt.hist(expZ, bins=np.arange(1, np.max(expZ)+2) - 0.5, density = True, alpha = 0.6, label="Empirique") 
plt.legend()
plt.show()

# Conclusion
print("L'esp√©rance math theorique de X est", 1/p_z)
print("L'esp√©rance math empirique de X est", np.mean(expZ))
print("L'ecart type thoerique de X est", np.sqrt((1-p_z)/p_z**2))
print("L'ecart type empirique de X est", np.std(expZ))
```
