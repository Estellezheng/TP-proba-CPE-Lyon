Simulation de la loi ùëã~ùê∫ùëíùëú(ùúÜ) avec ùúÜ = 0.3

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# theorie
lambda_x = 0.3  

# loi geometrique X
k = np.arange(1, 15)
geoX =  (1-lambda_x)**(k-1)*lambda_x     
plt.bar(k, geoX, color="red", label="Th√©orique loi de poisson")

# experimentation
essais = 100
expX =[]
for i in range(essais) :
    cpt = 0
    while np.random.rand() > lambda_x : 
        cpt += 1
    if cpt == 0 :
        cpt = 1
    expX.append(cpt)
    
    
def LoiBernoulli(p) :
    num = np.random.rand()
    if num < p :
        return 1 # succ√®s
    else :
        return 0 # √©chec

def LoiGeometrique(p) :
    events = 0
    while True :
        if LoiBernoulli(p) == 1 :
            events += 1
            break
        else : 
            events += 1
    return events

plt.hist(expX, bins=np.arange(1, 5) - 0.5, density = True, alpha = 0.6, label="Empirique") 
plt.legend()
plt.show()
```
