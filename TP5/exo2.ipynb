### Theoreme central limite

La densit√© de probabilit√© d‚Äôune somme de deux variables al√©atoires ind√©pendantes ùëã  et  ùëå,  de  densit√©s  ùëìùëã  et  ùëìùëå 
respectivement, s‚Äôobtient √† l‚Äôaide du produit de convolution : 
ùëìùëã+ùëå =ùëìùëã ‚àóùëìùëå 
Ce  r√©sultat  se  g√©n√©ralise  √†  un  nombre  quelconque  de  variables  al√©atoires  ind√©pendantes,  c‚Äôest-√†-dire  si  (ùëãùëò)1‚â§ùëò‚â§ùëõ  est  un ensemble de variables al√©atoires ind√©pendantes, et si on note ùëç=ùëã1 +ùëã2 +‚ãØ+ùëãùëõ alors : 
ùëìùëç =ùëìùëã1 ‚àóùëìùëã2 ‚àó...‚àóùëìùëãùëõ 

Dans le cas particulier o√π toutes les variables al√©atoires suivent la m√™me loi, alors la densit√© de ùëç est proche de celle d‚Äôune loi normale : c‚Äôest le th√©or√®me central limite. 

#### Histogramme de la loi

```python
import numpy as np
import matplotlib.pyplot as plt

# Loi de probabilit√© de X
K = np.arange(3,9)
Pk = (1/20)*np.array([4, 1, 3, 7, 2, 3])

# Affichage
plt.bar(K,Pk, label = 'Theorique', color = 'red')
plt.legend()

# moyennes et ecarts types  
print("La moyenne empirique Eemp(X) = ", np.sum(K*Pk))
print("L'ecart type empirique est oemp(X) = = ",np.sqrt(np.sum(K**2*Pk)- np.sum(K*Pk)**2) )
```

#### Loi de probabilit√© somme X1 + X1 <=> convolution

```python
# Somme
K = np.arange(2*3,2*8+1)    #(min, max)
conv = np.convolve(Pk, Pk)  # <=> somme

# Affichage
plt.bar(K, conv, label = 'loi de X1 + X2', color = 'blue')
plt.title('Histogramme de la somme de deux variables X1 et X2')
plt.legend()
```

#### Densit√© de X1 + X2 + .. + X10

```python
n = 10      #nmb de VA

# somme de 10 X
K = np.arange(n*3,n*8+1)
Z = Pk
for i in range(n-1) :
    Z = np.convolve(Z, Pk)
```

#### L'esperance et de l'ecart type

```python
# moyennes et ecarts types  
print("n =", n)
moyenne = np.sum(K*Z)
ecart = np.sqrt(np.sum(K**2*Z)- np.sum(K*Z)**2)
print("La moyenne empirique Eemp(X) = ", moyenne)
print("L'ecart type empirique est oemp(X) = = ",ecart) 
```

#### TCL : Theoreme centrale Limite

Tacrer la densit√© de la loi normale sous forme d‚Äôune  courbe donn√©e par le th√©or√®me central limite (sur le support [ùúá‚àí5ùúé ;ùúá+5ùúé] o√π ùúá et ùúé 

X1, X2, .. Xn sont des VA identiquement distribu√©es, ce sont des VA ind√©pendantes obeissant √† la meme loi d'esperance u et d'ecart type o

Le TCL stipule que, dans ces conditions, la somme et la moyenne de nombreuses VA ind√©pendantes et identiquement distribu√©es suivent une distribution normale, quelle que soit la distribution d'origine de ces variables.

Si on prend un grand nombre de ces variables et qu'on les additionne, la distribution de cette somme tendra √† √™tre normalement distribu√©e, m√™me si la distribution d'origine n'est pas normale.

Pour n variables, la moyenne de la somme est ùëõùúá et l'√©cart-type est 
sqrt(ùëõ*ùúé)
‚Äã

```python
# theorie
axe = np.linspace(moyenne - 5 * ecart, 80)
# loi normale 
densite_normale = (1 / (ecart * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((axe - moyenne) / ecart) ** 2)
```

### Affichage

```python
# affichage
plt.plot(axe, densite_normale, color='red', label='Densit√© th√©orique TCL', lw=2)
plt.bar(K, Z, label = 'loi de X1 + .. + X10', color = 'green')
plt.title('Histogramme de la somme de n = 10 variables al√©atoires X')
plt.legend()
```
