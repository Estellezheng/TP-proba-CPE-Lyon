Theoreme central limite

```python
import numpy as np
import matplotlib.pyplot as plt

K = np.arange(3,9)
Pk = (1/20)*np.array([4, 1, 3, 7, 2, 3])

plt.bar(K,Pk, label = 'Theorique', color = 'red')
plt.legend()

# moyennes et ecarts types  
print("La moyenne empirique Eemp(X) = ", np.sum(K*Pk))
print("L'ecart type empirique est oemp(X) = = ",np.sqrt(np.sum(K**2*Pk)- np.sum(K*Pk)**2) )
```

B. Loi de probabilité somme X1 + X1 <=> convolution

```python
K = np.arange(2*3,2*8+1)
conv = np.convolve(Pk, Pk)

plt.bar(K, conv, label = 'loi de X1 + X2', color = 'blue')
plt.legend()
```

C. Densité de X1 + .. + X10

```python
n = 10      #nmb de VA

# methode de convolution
K = np.arange(n*3,n*8+1)
Z = Pk
for i in range(n-1) :
    Z = np.convolve(Z, Pk)

# affichage
plt.bar(K, Z, label = 'loi de X1 + .. + X10', color = 'green')
plt.legend()
```

D. Affichage de l'esperance et de l'ecart type

```python
# moyennes et ecarts types  
print("n =", n)
print("La moyenne empirique Eemp(X) = ", np.sum(K*Z))
print("L'ecart type empirique est oemp(X) = = ",np.sqrt(np.sum(K**2*Z)- np.sum(K*Z)**2) )
```
