### TP2 - EXO2 - Simulations des lois de probabilité usuelles discrètes

Fonctions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
import math 

def LoiUniform(n) :
    return np.random.randint(1, n+1)

def LoiBernouilli(p):
    nb = np.random.rand()
    if nb < p : return 1
    else : return 0

def LoiBinomiale(n,p) :
    cpt = 0
    for i in range(n):
        if LoiBernouilli(p) : 
            cpt += 1 
    return cpt

def LoiGeometrique(p):
    cpt = 1
    while (LoiBernouilli(p)==0) :
        cpt += 1
    if cpt == 0 : cpt = 1
    return cpt
```

# Test Loi Uniforme

```python
n = 6

#theorie
K = np.arange(1,n+1)   
probth = []

for k in range(n) : 
    probth.append(1/n)
plt.bar(K, probth, color="red", label="Théorique")

# empirique
essais = 10000
uniforme =[]
for i in range(essais) :
    uniforme.append(LoiUniform(n))
plt.hist(uniforme, bins=np.arange(1, n+1) - 0.5, density = True, alpha = 0.6, label="Empirique")  

plt.legend()
plt.show()

print("L'espérance math theorique de X est", (n+1)/2)
print("L'espérance math empirique de X est", np.mean(uniforme))
print("L'ecart type empirique de X est", np.std(uniforme))
```

# Test Loi Bernouilli

```python
p = 0.28

#theorie
K = np.arange(0,2)   
probth = [1-p, p]
plt.bar(K, probth, color = "red", label="Théorique")

# empirique
essais = 100
bernouilli =[]
for i in range(essais) :
    bernouilli.append(LoiBernouilli(p))
plt.hist(bernouilli, bins=np.arange(0, 5) - 0.5, density = True, alpha = 0.6, label="Empirique") 

plt.legend()
plt.show()

print("L'espérance math theorique de X est", p)
print("L'espérance math empirique de X est", np.mean(bernouilli))
print("L'ecart type empirique de X est", np.std(bernouilli))
```

# Test Loi Binomiale

```python
n = 40
p = 0.28

#theorie
K = np.arange(0,n+1)   
probth = []

for k in range(n+1) : 
    combinaison = math.factorial(n)/(math.factorial(k)*math.factorial(n-k))
    x = combinaison*p**k*(1-p)**(n-k)
    probth.append(x)

plt.bar(K, probth, color = "red", label="Théorique")

# empirique
essais = 10000
binomiale =[]
for i in range(essais) :
    binomiale.append(LoiBinomiale(n,p))
plt.hist(binomiale, bins=np.arange(0, n+1) - 0.5, density = True, alpha = 0.6, label="Empirique")  

plt.legend()
plt.show()

print("L'espérance math theorique de X est", n*p)
print("L'espérance math empirique de X est", np.mean(binomiale))
print("L'ecart type empirique de X est", np.std(binomiale))
```

# Test Loi geometrique

```python
p = 0.2
n = 10

#theorie
K = np.arange(1,n+1)      
probth = []

for k in range (1,n+1) : 
    probth.append((1-p)**(k-1)*p)
plt.bar(K, probth, color = "red", label = "Théorique")


# empirique
essais = 1000
geometrique =[]

for i in range(essais) :
    geometrique.append(LoiGeometrique(p))
plt.hist(geometrique, bins=np.arange(1, 10) - 0.5, density = True, alpha = 0.5, label = "Empirique")

plt.legend()
plt.show()

print("L'espérance math theorique de X est", 1/p)
print("L'espérance math empirique de X est", np.mean(geometrique))
print("L'ecart type empirique de X est", np.std(geometrique))
```
