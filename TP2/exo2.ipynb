### Definition des lois de proba

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
import math 

# retourne un nombre entier aléatoire entre 0 et n
def LoiUniforme(n) :
    return np.random.randint(1, n+1)

# génère un nombre aléatoire entre 0 et 1, retourne 1 si ce nombre est inférieur à p ou 0 dans le cas contraire
def LoiBernouilli(p):
    nb = np.random.rand()
    if nb < p : return 1
    else : return 0

#retourne  le  nombre  de  succès  d'un  évènement  de  probabilité  p  répété  n  fois
def LoiBinomiale(n,p) :
    cpt = 0
    for i in range(n):
        if LoiBernouilli(p) : 
            cpt += 1 
    return cpt

#etourne le nombre d'évènements ayant eu lieu avant que le 1er succès de probabilité p ne se produise 
def LoiGeometrique(p):
    cpt = 0
    while (LoiBernouilli(p)==0) :
        cpt += 1
    return cpt + 1      #il y a k-1 echecs et le kè succes
```

### Test Loi Uniforme

```python
n = 6

#theorie
K = np.arange(1,n+1)   
probth = [1/n] * n
plt.bar(K, probth, color="red", label="Théorique")

# empirique
essais = 10000
uniforme = [LoiUniforme(n) for i in range(essais)]
#uniforme =[]
#for i in range(essais) :
#    uniforme.append(LoiUniforme(n))
plt.hist(uniforme, bins=np.arange(1, n+2) - 0.5, density = True, alpha = 0.6, label="Empirique")  

plt.legend()
plt.show()

# Conclusion
print("L'espérance math theorique de X est", (n+1)/2)
print("L'espérance math empirique de X est", np.mean(uniforme))
print("L'ecart type theorique de X est", np.sqrt((n**2-1)/12))
print("L'ecart type empirique de X est", np.std(uniforme))
```

### Test Loi Bernouilli

```python
p = 0.28

#theorie
K = np.arange(0,2)   
probth = [1-p, p]
plt.bar(K, probth, color = "red", label="Théorique")

# empirique
essais = 100
bernouilli =[LoiBernouilli(p) for i in range(essais)]
plt.hist(bernouilli, bins=np.arange(0, 3) - 0.5, density = True, alpha = 0.6, label="Empirique") 

plt.legend()
plt.show()

print("L'espérance math theorique de X est", p)
print("L'espérance math empirique de X est", np.mean(bernouilli))
print("L'ecart type theorique de X est", np.sqrt(p*(1-p)))
print("L'ecart type empirique de X est", np.std(bernouilli))
```

### Test Loi Binomiale

```python
n = 40
p = 0.28

#theorie
K = np.arange(0,n+1)   
probth = [(math.factorial(n)/(math.factorial(k)*math.factorial(n-k)))*(p**k)*((1-p)**(n-k)) for k in K]
plt.bar(K, probth, color = "red", label="Théorique")

# empirique
essais = 10000
binomiale =[LoiBinomiale(n,p) for i in range(essais)]
plt.hist(binomiale, bins=np.arange(0, n+2) - 0.5, density = True, alpha = 0.6, label="Empirique")  

plt.legend()
plt.show()

print("L'espérance math theorique de X est", n*p)
print("L'espérance math empirique de X est", np.mean(binomiale))
print("L'ecart type empirique de X est", np.sqrt(n*p*(1-p)))
print("L'ecart type empirique de X est", np.std(binomiale))
```

# Test Loi geometrique

```python
p = 0.2
n = 10

#theorie
K = np.arange(1,n+1)      
probth = (1-p)**(K-1)*p
plt.bar(K, probth, color = "red", label = "Théorique")

# empirique
essais = 1000
geometrique =[LoiGeometrique(p) for i in range(essais)]
plt.hist(geometrique, bins=np.arange(1, n+2) - 0.5, density = True, alpha = 0.5, label = "Empirique")

plt.legend()
plt.show()

print("L'espérance math theorique de X est", 1/p)
print("L'espérance math empirique de X est", np.mean(geometrique))
print("L'ecart type thoerique de X est", np.sqrt((1-p)/p**2))
print("L'ecart type empirique de X est", np.std(geometrique))
```
