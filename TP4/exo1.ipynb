Simulation loi uniforme sur [a,b]

```python
import numpy as np
import matplotlib.pyplot as plt

a = -2
b = 3
axe = np.linspace(a,b,30)

# Simulation empirique de la loi uniforme
print("X est la VA qui suit la loi uniforme sur [a,b]")
X = np.random.rand(10000)*(b-a)+a  

# densité de proba
theo = []
for i in axe : 
    theo.append(1/(b-a))

# affichage
hX_prob, binsX_prob, _ = plt.hist(X, bins=30, density=True, color = "green",edgecolor='black', label = 'Empirique')
plt.xlabel('valeurs de X')
plt.title('Simulation de la loi uniforme sur l intervalle [-2,3]')

plt.plot(axe,theo, label = 'Theorique')
plt.legend()

# moyennes et ecarts types
print("La moyenne théorique Eth(X) = ", (a+b)/2)
print("La moyenne empirique Eemp(X) = ", np.mean(X))
print("L'ecart type théorique est oth(X) =", np.sqrt((b-a)**2/12))
print("L'ecart type empirique est oemp(X) = = ", np.std(X))
```
