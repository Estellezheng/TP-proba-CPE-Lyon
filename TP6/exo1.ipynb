```python
import numpy as np
import matplotlib.pyplot as plt

n = 50000
axe = np.linspace(0,15,100)

# Simulation empirique de la loi uniforme
# X est une loi exponentielle avec lambda = 1
Fx = np.random.rand(n)
X = -np.log(1-Fx)
# Y est une loi exponentielle avec lambda = 2
Fy = np.random.rand(n)
Y = -np.log(1-Fy)/2
# definition de U
U = X + 2*Y

# fonction theorique
theo = axe*np.exp(-axe)

# Affichage de U
plt.plot(axe,theo, label = 'Theorique', color = 'red')
plt.legend()

hX_prob, binsX_prob, _ = plt.hist(U, bins=30, density=True, color = "green",edgecolor='black', label = 'Empirique')
plt.title('Simulation de la loi ue^-u')
plt.legend()

# moyennes et ecarts types
print("Le nombre d'essai est = ", n)
print("La moyenne théorique Eth(X) = ", 2)    #d'après le td
print("La moyenne empirique Eemp(X) = ", np.mean(U))
print("La variance théorique est oth(X) =", 2)
print("La variance empirique est oemp(X) =", np.var(U))

```
