```python
import numpy as np
import matplotlib.pyplot as plt

n = 50000

# fonction theorique
axe = np.linspace(0,15,100)
theo = axe*np.exp(-axe)     #TD fonction de densit√© de U
plt.plot(axe,theo, label = 'Theorique', color = 'red')
plt.legend()

# Simulation empirique

# Les lois exponentielles sont simul√©es par methode d'inversion

# X est une loi exponentielle avec lambda = 1
U1 = np.random.rand(n)
# U1 = FX(X) = 1-exp(-lambda*X)
# exp(-lambda*X) = 1 - U1
# -lambda*X = ln(1-U1)
# X = -ln(1-U1)/lambda
# X = -ln(1-U1)
X = -np.log(1-U1)   # X ~ Exp(1)

# Y est une loi exponentielle avec lambda = 2
U2 = np.random.rand(n)
# -lambda*Y = ln(1-U2)
# X = -ln(1-U1)/2
Y = -np.log(1-U2)/2     # Y ~ Exp(2)

# Definition de U
U = X + 2*Y

# Affichage de U
hX_prob, binsX_prob, _ = plt.hist(U, bins=30, density=True, color = "green",edgecolor='black', label = 'Empirique')
plt.title('Simulation de la loi ue^-u')
plt.legend()

# moyennes et ecarts types
print("Le nombre d'essai est = ", n)
print("La moyenne th√©orique Eth(X) = ", 2)    #d'apr√®s le td
print("La moyenne empirique Eemp(X) = ", np.mean(U))
print("La variance th√©orique est oth(X) =", 2)
print("La variance empirique est oemp(X) =", np.var(U))

```
