## Partie 1 : lois unifome

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100000
a = 0
b = 1

# X et Y suivent une loi uniforme sur [0,1]
X = np.random.rand(n)*(b-a) + a
Y = np.random.rand(n)*(b-a) + a
```

```python
# Affichage des histogrammes empiriques pour X et Y
plt.hist(X, bins=50, density=True,edgecolor='black', alpha=0.6, label='X')
plt.hist(Y, bins=50, density=True,edgecolor='black', alpha=0.6, label='Y')
plt.xlabel('x et y')
plt.legend() 
```

```python
#calcul de Z 
Z = np.minimum(X, Y)
plt.hist(Z, bins=50, density=True, edgecolor='black',alpha=0.6, label='Zemp')
plt.xlabel('z')
plt.legend()

# Loi thorique
axe = np.arange(0, b, 0.01)
theo = 2 * (1 - axe)    # d'apres le td exo 5
plt.plot(axe, theo, 'b', label='zth (Loi Thorique)')
plt.xlabel('z')
plt.legend()

plt.show()
```

## Partie  2: X et Y suivent une loi exponentielle

```python
# Simulation empirique pour exponentielle
lambda_valeur = 1
A_Xemp = np.random.rand(n)
Xemp = -np.log(1 - A_Xemp) /lambda_valeur

A_Yemp = np.random.rand(n)
Yemp = -np.log(1 - A_Yemp) /lambda_valeur
```

```python
# Affichage des histogrammes empiriques pour X et Y
plt.hist(Xemp, bins=50, density=True,edgecolor='black', alpha=0.6, label='Xemp')
plt.hist(Yemp, bins=50, density=True,edgecolor='black', alpha=0.6, label='Yemp')
plt.xlabel('x et y')
plt.legend()
```

```python
#calcul de Zemp 
Zemp = np.minimum(Xemp, Yemp)
plt.hist(Zemp, bins=50, density=True, edgecolor='black',alpha=0.6, label='Zemp')
plt.xlabel('z')
plt.legend()

# Loi thorique
theo = 2*lambda_valeur*np.exp(-2*lambda_valeur*axe) #densite theorique ; f z de z 

plt.plot(axe, theo, 'b', label='zth (Loi Thorique)')
plt.xlabel('z')
plt.legend()

plt.show()
```
