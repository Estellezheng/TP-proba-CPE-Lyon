## Cas 1 : X et Y suivent une lois uniforme

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100000
a = 0
b = 1

# X et Y suivent une loi uniforme sur [0,1]
X = np.random.rand(n)
Y = np.random.rand(n)

# calcul de Z 
Z = np.minimum(X, Y)
plt.hist(Z, bins=50, density=True, edgecolor='black',alpha=0.6, label='Zemp')
plt.xlabel('z')
plt.legend()

# Loi théorique
axe = np.arange(a, b, 0.01)
theo = 2 * (1 - axe)    # d'apres le td exo 5
plt.plot(axe, theo, 'b', label='zth (Loi Théorique)')
plt.xlabel('z')
plt.legend()

plt.show()
```

## Cas  2 : X et Y suivent une loi exponentielle

```python
# Simulation empirique pour exponentielle
lambda_valeur = 1
U1 = np.random.rand(n)
Xemp = -np.log(1 - U1) /lambda_valeur

U2 = np.random.rand(n)
Yemp = -np.log(1 - U2) /lambda_valeur

#calcul de Z
Zemp = np.minimum(Xemp, Yemp)
plt.hist(Zemp, bins=50, density=True, edgecolor='black',alpha=0.6, label='Zemp')
plt.xlabel('z')
plt.legend()

# Loi théorique
theo = 2*lambda_valeur*np.exp(-2*lambda_valeur*axe)     # TD fz de z 
plt.plot(axe, theo, 'b', label='Théorique')
plt.xlabel('z')
plt.legend()

plt.show()
```
