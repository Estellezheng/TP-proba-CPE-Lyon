Proposer une m√©thode qui permet de g√©n√©rer deux variables al√©atoires corr√©l√©es avec un coefficient de corr√©lation donn√©. 
1)  [calcul  pr√©liminaire  ¬´ √†  la  main ¬ª]  Soient  ùëã  et  ùëå  deux  variables  al√©atoires  ind√©pendantes  distribu√©es  selon  une  loi normale  centr√©e r√©duite. Soit  ùëç =ùëéùëã+ùëèùëå, montrer que  ùëâ(ùëç)=ùëé2 +ùëè2  puis en d√©duire  le coefficient de  corr√©lation 
du couple (ùëã,ùëç) puis du couple (ùëå,ùëç). 
2)  V√©rifier de fa√ßon empirique le r√©sultat pr√©c√©dent dans le cas o√π le coefficient de corr√©lation du couple (ùëã,ùëç) est √©gal √† ‚àí‚àö3/2 avec la contrainte suppl√©mentaire que l‚Äô√©cart type de ùëç soit √©gal √† 1. Pour cela : 

- tracer le nuage de point  (ùëã,ùëå) (utiliser la fonction matplotlib.pyplot.scatter, comparer les coefficients de 
corr√©lation empirique et th√©orique (utiliser la fonction numpy.corrcoef) et interpr√©ter les r√©sultats 

- m√™mes questions avec les couples (ùëã,ùëç) et (ùëå,ùëç

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000
axe = np.linspace(0,10,100)

# Simulation empirique de la loi normale centr√©e reduite pour X et Y
X = np.random.randn(n)      # randn pour une loi normale centree reduite
Y = np.random.randn(n)  

#cov(X,Z) = a et cov(Y,Z) = b par calcul
# definition de Z
a = -np.sqrt(3)/2   #coef de correlation
b = np.sqrt(1 - a**2)            # conditions : V(Z) = a¬≤ + b¬≤ = 1 : b = sqrt(1-a¬≤) et ecart type de Z = 1
Z = a*X + b*Y
```

### Affichage XZ

```python
# affichage nuage de points
plt.scatter(X,Z, color = 'blue')
plt.axis ='equal'
plt.title("Nuage de points (X, Z)")
plt.xlabel("X")
plt.ylabel("Z")
plt.grid()

# conclusion
# matrice de variance emp = np.corrcoef(X,Y) = [[V(X), cov(X,Y)],[cov(X,Y), V(Y)]]
# calcul coef_correlation = cov(X,Y)/(sqrt(V(X))*sqrt(V(Y)))
print("Le coefficient de correlation de X, Z est :", np.corrcoef(X,Z)[0][1]/(np.sqrt(np.corrcoef(X,Z)[0][0])*np.sqrt(np.corrcoef(X,Z)[1][1])))
print("Le coefficient theorique de correlation de Y,Z est ", a)
```

### Affichage YZ

```python
# affichage nuage de points
plt.scatter(Y,Z, color = 'green') 
plt.axis ='equal'
plt.title("Nuage de points (Y, Z)")
plt.xlabel("Y")
plt.ylabel("Z")
plt.grid()

# conclusion
print("Le coefficient de correlation de X, Z est :", np.corrcoef(Y,Z)[0][1]/(np.sqrt(np.corrcoef(Y,Z)[0][0])*np.sqrt(np.corrcoef(Y,Z)[1][1])))  
print("Le coefficient theorique de correlation de Y,Z est ", b/np.sqrt(a**2+b**2))
```

```python
# affichage nuage de points
plt.scatter(X,Y, color = 'red')   
plt.axis ='equal'
plt.title("Nuage de points (X, Y)")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()

# conclusion
print("Le coefficient de correlation de X, Z est :", np.corrcoef(X,Y)[0][1]/(np.sqrt(np.corrcoef(X,Y)[0][0])*np.sqrt(np.corrcoef(X,Y)[1][1])))  
print("Le coefficient theorique de correlation de X, Y est 0")
```
