Exercice 2 - générer deux variables aléatoires corrélées 

```python
import numpy as np
import matplotlib.pyplot as plt

n = 1000
axe = np.linspace(0,10,100)

# Simulation empirique de la loi normale centrée reduite 
print("X et Y sont des VA qui suivent la loi normale")
X = np.random.randn(n)      # randn pour une loi normale centree reduite
Y = np.random.randn(n)  

#cov(X,Z) = a et cov(Y,Z) = b par calcul
# definition de Z
a = -np.sqrt(3)/2
b = -1/2   # conditions : V(Z) = a² + b² = 1 et ecart type de Z = 1
Z = a*X + b*Y
```

```python

# affichage nuage de points
plt.scatter(X,Z, color = 'blue')
plt.axis ='equal'

# theorie
print("Le coefficient de correlation de X, Z est :", np.corrcoef(X,Z)[0][1]/(np.sqrt(np.corrcoef(X,Z)[0][0])*np.sqrt(np.corrcoef(X,Z)[1][1])))
print("Le coefficient theorique de correlation de Y,Z est ", np.sqrt(3)/2)
```

```python

# affichage nuage de points
plt.scatter(Y,Z, color = 'green') 

# theorie
print("Le coefficient de correlation de X, Z est :", np.corrcoef(Y,Z)[0][1]/(np.sqrt(np.corrcoef(Y,Z)[0][0])*np.sqrt(np.corrcoef(Y,Z)[1][1])))  
print("Le coefficient theorique de correlation de Y,Z est ", b/np.sqrt(a**2+b**2))
```
